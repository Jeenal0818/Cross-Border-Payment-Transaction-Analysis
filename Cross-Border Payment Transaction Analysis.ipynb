```python
from crewai import Agent, Task, Crew
from crewai import LLM
```

```python
llm = LLM(model="gpt-4o-mini", temperature=0.0)
```

```python
payment_analyzer = Agent(
    role="Payment Analyzer",
    goal="Strictly analyze cross-border payment transactions based only on the provided inputs.",
    backstory="Expert in payment systems, fraud detection, and compliance. Never assume details not provided.",
    llm=llm
)
```

```python
analyze_payment_task = Task(
    description=(
        "You are given a cross-border payment transaction with these details:\n"
        "Amount: {amount}\n"
        "Origin Country: {origin[country]}\n"
        "Origin Currency: {origin[currency]}\n"
        "Origin Bank: {origin[bank]}\n"
        "Destination Country: {destination[country]}\n"
        "Destination Currency: {destination[currency]}\n"
        "Destination Bank: {destination[bank]}\n\n"
        "Analyze ONLY this transaction. Do not assume or add extra details. "
        "Return the result strictly in JSON format with keys:\n"
        "sender_country, receiver_country, amount, currency, origin_bank, destination_bank, risk_score, reasoning."
    ),
    expected_output="JSON",
    agent=payment_analyzer
)
```

```python
from crewai import Task, Crew

def build_payment_task(inputs):
    return Task(
        description=(
            f"You are given a cross-border payment transaction with these details:\n"
            f"- Amount: {inputs['amount']} {inputs['origin']['currency']}\n"
            f"- Origin Country: {inputs['origin']['country']}\n"
            f"- Origin Bank: {inputs['origin']['bank']}\n"
            f"- Destination Country: {inputs['destination']['country']}\n"
            f"- Destination Bank: {inputs['destination']['bank']}\n\n"
            "Analyze ONLY this transaction. Do not assume or add extra details. "
            "Return the result strictly in Markdown format with clear headings and bullet points."
        ),
        expected_output="Markdown",
        agent=payment_analyzer
    )
```

```python
inputs = {
    "amount": 10000,
    "origin": {"country": "India", "currency": "INR", "bank": "HDFC Bank"},
    "destination": {"country": "USA", "currency": "USD", "bank": "Bank of America"}
}

# Create the analysis task
analyze_payment_task = build_payment_task(inputs)

crew = Crew(
    agents=[payment_analyzer],
    tasks=[analyze_payment_task],
    process="sequential"
)

result = crew.kickoff(inputs=inputs)

print(result)
```
